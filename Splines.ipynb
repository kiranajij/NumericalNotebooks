# Linear Spline

Given interpolating points $x_0, x_1, \cdots, x_n$ we have the points $y_0, y_1, \cdots, y_n$. Then linear spline of is defined by 

$$
    P(x) = \frac{x-x_{i}}{x_{i-1}-x_{i}}y_{i-1} + \frac{x-x_{i-1}}{x_i-x_{i-1}} y_i \\
    x\in[x_{i-1}, x_i]
$$


```python
def find_min_i(xx, Xi):
    # Can be done using Binary Search but meh
    for i in range(len(Xi)):
        if xx <= Xi[i]:
            return i-1

class LinearSpline:
    def __init__(self, Xs, Ys):
        self.Xs = Xs
        self.Ys = Ys
    
    def __call__(self, xx):

        Xs = self.Xs
        Ys = self.Ys
        
        if xx < Xs[0] or x > Xs[-1]:
            return 0
        i = find_min_i(xx, self.Xs)
#         print(i, xx)
        
        l2 = (xx-Xs[i])/(Xs[i+1]-Xs[i])*Ys[i+1]
        l1 = (xx-Xs[i+1])/(Xs[i]-Xs[i+1])*Y[i]
        
        return N(l1+l2)
    def __lfloat__(self):
        print("called")
```
