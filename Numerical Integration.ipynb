# Euler Method


```python
# Problem statement
# y'=x+y
# y(0) = 1

f(x,y) = x^2+y
y0 = -2; x0=0
phi(x) = -x^2-2*x-2

h = 0.1
xs = srange(x0, 4, h)
ys = [y0]

for i in range(1,(len(xs))):
    ynn = ys[i-1] + h*f(xs[i-1], ys[i-1])
    ys.append(ynn)
    
    
p = plot(phi, (x, 0, 4), linestyle='--', thickness=2, legend_label='exact solution')
p +=line2d(zip(xs, ys), color='red', legend_label='Euler method')
p += scatter_plot(zip(xs, ys), facecolor='black', markersize=10)
show(p)
```

```python
def solve_RK4(xs, f, y0):
    
    h = xs[1]-xs[0]
    ys = [y0]
    
    for i in range(1, len(xs)):
        xn = xs[i-1]
        yn = ys[i-1]
        
        k1 = f(xn, yn)
        k2 = f(xn+h/2, yn+h*k1/2)
        k3 = f(xn + h/2, yn+h*k2/2)
        k4 = f(xn+h, yn+h*k3)
        
        ynn = yn + h*(k1+2*k2+2*k3+k4)/6
        ys.append(ynn)
    return ys
ys = solve_RK4(xs, f, y0)
p = plot(phi, (x, 0, 4), linestyle='--', thickness=2, legend_label='exact solution')
p +=line2d(zip(xs, ys), color='red', legend_label='RK4')
p += scatter_plot(zip(xs, ys), facecolor='black', markersize=10)
show(p)
```
